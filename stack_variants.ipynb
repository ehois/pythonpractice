```python
class Node:
    """A generic Node class.
    
    The purpose is to provide a generic class, therefore the specific setting of next is up to the data structure 
    and usually done when a Node is added or removed there.

    Attributes:
        item : Holds the actual content of the node.
        next : Points to next Node according to the data structure it's used in. Defaults to None.

    """
    def __init__(self, item):
        self.item = item
        self.next = None
    
    def __str__(self): 
        """Specifies what is used when using print() on a Node.
        
        Returns:
            item: The item, so the content of the Node will be printed on print().

        """
        return str(self.item)

class Stack:
    """A simple Stack class (LIFO).
    
    Attributes:
        head : Points to the top element of the stack. Defaults to None at the start and when stack is empty.
        size : Counts number of elements of the stack. Initialized with 0.

    """
    
    def __init__(self):
        self.head = None
        self.size = 0
        
        # TODO: implement a method isEmpty to check instead of making size public? (also need one method to change size then)
        # Then we would use self.isEmpty as check instead of self.size == 0 later
    
    def push(self, item): 
        """Creates a Node for the given item and puts in on the stack.

        Args:
            item: The content to be put on the stack.

        """
        
        # create node for new element
        node = Node(item)
        
        node.next = self.head
        
        # check if stack is empty (otherwise there is no head for next step and new node is just set as head)
        if self.size == 0:
            self.head = node
            
        self.head = node
        self.size += 1
    
    def pop(self):
        """Returns top of the stack.

        Returns:
            Node: The current head of the stack.

        """
        
        
        # check if stack is empty first
        if self.size == 0:
            print("Stack is Empty")
            return None
        
        # temporarily save head in new variable to be returned later
        old_head = self.head
        self.head = old_head.next
        self.size -=1

        return old_head
    
    def __str__(self):
        """Specifies what is used when using print() on a Stack.

        Returns:
            list: the current items of the Nodes on the stack are added to a list and returned.

        """
        if self.size == 0:
            return str(list())
            
        l = list()
        #TODO: Don't like, that we need to know how Node looks like (we name item explicitly, avoidable?)
        l.append(self.head.item)
        
        n = self.head.next
        while (n != None):
            l.append(n.item)
            n = n.next
        return str(l)

    
s = Stack()
s.push(4)
s.push(6)
s.push(8)
print(s)


x = s.pop()
print(s)

x = s.pop()
print(s)

x = s.pop()
print(s)

x = s.pop()
print(s)



#print(Node.__doc__)
#print(Stack.__doc__)

        
        
        
    
```
